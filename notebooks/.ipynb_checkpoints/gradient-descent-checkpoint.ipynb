```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def L(x,y):
    return (x-1)**2 + (y-2)**2 - 1*x*y
```

```python
delta = 0.1
x = np.arange(-10.0, 10.0, delta)
y = np.arange(-10.0, 10.0, delta)
X, Y = np.meshgrid(x, y)

Z = []
for xx in x:
    for yy in y:
        Z.append(L(xx,yy))
Z = np.array(Z).reshape(X.shape[0],X.shape[1])

fig, ax = plt.subplots(figsize = (10,7))
CS = ax.contourf(X, Y, Z)
cbar = fig.colorbar(CS)
ax.set_xlabel(r"$\beta_1$", fontsize = 16)
ax.set_ylabel(r"$\beta_2$", fontsize = 16)
plt.show()
```

```python
def gradient_L(beta):
    return np.array([2*(beta[0]-1) - beta[1],
            2*(beta[1]-2)-beta[0]])

beta0 = np.array([-5.0, -2.5])
beta_sol = beta0
gamma = 0.68
beta_visited=[]
for i in range(25):
    beta_visited.append(beta_sol)
    beta_sol = beta_sol - gamma*gradient_L(beta_sol)
    #print(beta_sol)
    
delta = 0.1
x = np.arange(-10.0, 10.0, delta)
y = np.arange(-10.0, 10.0, delta)
X, Y = np.meshgrid(x, y)

Z = []
for xx in x:
    for yy in y:
        Z.append(L(xx,yy))
Z = np.array(Z).reshape(X.shape[0],X.shape[1])

fig, ax = plt.subplots(figsize = (10,7))
CS = ax.contourf(X, Y, Z)
cbar = fig.colorbar(CS)
ax.set_xlabel(r"$\beta_1$", fontsize = 16)
ax.set_ylabel(r"$\beta_2$", fontsize = 16)
for bb in beta_visited:
    plt.scatter(bb[0], bb[1], color ="red", s=50)
plt.scatter(8/3, 10/3, color ="green", marker ='*', s=200)
plt.show()
```
