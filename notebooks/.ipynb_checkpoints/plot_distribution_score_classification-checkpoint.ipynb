```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gauss(x, mu, sigma):
    return (1/(sigma*np.sqrt(6.28)))*np.exp(-(x-mu)**2/(2*sigma**2))

x = np.arange(0,1,0.01)
gauss_1 = [ gauss(xx, 0.35, 0.15) for xx in x]
gauss_2 = [ gauss(xx, 0.55, 0.2) for xx in x]

fig, ax = plt.subplots(1,1,figsize=(10,5))
ax.plot(x, gauss_1, label="$f_0 = p(score|y=0)$", color="red")
ax.plot(x, gauss_2, label="$f_1 = p(score|y=1)$", color="green")
ax.plot([0.3,0.3],[0,max(max(gauss_1),  max(gauss_2))+0.1], "--", c="blue", label="thr")
ax.set_xlabel("score", fontsize=16)
ax.set_ylabel("distribution", fontsize=16)
ax.legend(fontsize=14)
plt.show()
```
